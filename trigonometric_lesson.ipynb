```python
import numpy as np
import matplotlib.pyplot as plt

def my_sin(x):
    return np.sin(x)

def my_cos(x):
    return np.cos(x)

x = np.linspace(-10, 10)
y_sin = my_sin(x)
y_cos = my_cos(x)

plt.plot(x, y_sin, label="sin")
plt.plot(x, y_cos, label="cos")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def my_tan(x):
    return np.tan(x)

x = np.linspace(-np.pi, np.pi)
y_tan = my_tan(x)

plt.plot(x, y_tan, label="tan")
plt.legend()
plt.show()
```
