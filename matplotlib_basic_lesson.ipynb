```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10)
y = np.linspace(-10, 10)

y_1 = x
y_2 = x * x

plt.xlabel("x value")
plt.ylabel("y value")

plt.title("My Graph")

plt.plot(x, y_1, label="y1")
plt.plot(x, y_2, label="y2")
plt.legend()

plt.show()
```
